```python
from keras.preprocessing.image import ImageDataGenerator
from keras.models import load_model
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load the model we saved
model = load_model('cnn_model_weights_12c.h5')
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# predicting images
img = image.load_img('./dataset/test_set/rice_cooker/81571.jpg', target_size=(128, 128))
img = image.load_img('./dataset/test_set/frying_pan/37981.jpg', target_size=(128, 128))
img = image.load_img('./dataset/test_set/wine_glass/18135.jpg',  target_size=(128, 128))
img = image.load_img('./dataset/test_set/microwave/57943.jpg', target_size=(128, 128))
img = image.load_img('./dataset/test_set/pots/1286.jpg',  target_size=(128, 128))
img = image.load_img('./dataset/test_set/chopper/64879.jpg', target_size=(128, 128))
img = image.load_img('./dataset/test_set/swing/48083.jpg', target_size=(128, 128))
img = image.load_img('./dataset/test_set/wine_cooler/70971.jpg',  target_size=(128, 128))
img = image.load_img('./dataset/test_set/stools/7341.jpg', target_size=(128, 128))
img = image.load_img('./dataset/test_set/scissors/72132.jpg',  target_size=(128, 128))
img = image.load_img('./dataset/test_set/massage_chair/108541.jpg',  target_size=(128, 128))
# img = image.load_img('./dataset/test_set/clock/28010.jpg', target_size=(128, 128))

label=['clock','massage_chair','microwave','swing','stools','rice_cooker','frying_pan','chopper', 'pots','scissors','wine_cooler','wine_glass']

x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
images = np.vstack([x])
```

```python
classes = model.predict_classes(images)
plt.figure(figsize=(5,5))
plt.imshow(img)
# plots(images)  
print('Predict below item is a : ', label[classes[0]])
```
